```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/sample_data/tmdb_5000_movies.csv',delimiter=',')
credits = pd.read_csv('/content/sample_data/tmdb_5000_credits.csv',delimiter=',')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merge movies dataset with credits dataset into one dataset
movie_data = movies.merge(credits,on='title')
```

```python
movie_data.head(1)
```

```python
#for movie recommendations system below are the important parameters
#genres, movie_id, kewords, title, overview, cast, crew

movie_data = movie_data[['movie_id','title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movie_data.head()
```

```python
movie_data.isnull().sum()
```

```python
movie_data.dropna(inplace=True)
```

```python
movie_data.duplicated().sum()
```

```python
movie_data.iloc[0].genres
```

```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
#convert this into
#['Action','Adventure','Fantasy','SciFi']
```

```python
import ast
def convert(obj):
  List = []
  for i in ast.literal_eval(obj):
    List.append(i['name'])
  return List
```

```python
movie_data['genres'] = movie_data['genres'].apply(convert)
movie_data['keywords'] = movie_data['keywords'].apply(convert)
```

```python
movie_data.head()
```

```python
import ast

def convert3(obj):
  List = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      List.append(i['name'])
      counter+=1
    else:
      break
  return List
```

```python
movie_data['cast'] = movie_data['cast'].apply(convert3)
```

```python
movie_data.head()
```

```python
def fetch_director(obj):
  List = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
     List.append(i['name'])
     break
  return List
```

```python
movie_data['crew']= movie_data['crew'].apply(fetch_director)
```

```python
movie_data.head()
```

```python
movie_data['overview'][0]
```

```python
movie_data['overview'] = movie_data['overview'].apply(lambda x:x.split())
```

```python
movie_data.head()
```

```python
movie_data['genres'] = movie_data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie_data['keywords'] = movie_data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie_data['cast'] = movie_data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie_data['crew'] = movie_data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data.head()
```

```python
movie_data['tags'] = movie_data['overview'] + movie_data['genres'] + movie_data['keywords'] + movie_data['cast'] + movie_data['crew']
movie_data.head()
```

```python
new_df = movie_data[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df.head()
```

Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
recommend('Iron Man')
```
